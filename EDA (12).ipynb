```python
lap['OpSys']=lap['OpSys'].replace('No','Windows')
#replaced with Windows for in the No value 
```

```python
#prepare the data 
x=lap[['Company', 'TypeName', 'Ram', 'OpSys', 'Weight', 'Touchscreen','HDD','SSD','GPU']]
y=lap['Price'] #Dependent variable(target)
```

```python
x
```

Model Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_test
```

Feature Encoding 

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,median_absolute_error,mean_absolute_error
from sklearn.pipeline import Pipeline

step1=ColumnTransformer(transformers=[
    ('col_trn',OneHotEncoder(drop='first'),[0,1,3,8])
],remainder='passthrough')

step2=LinearRegression()

pip=Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pip.fit(x_train,y_train)

y_pred=pip.predict(x_test)

print('r2_score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
